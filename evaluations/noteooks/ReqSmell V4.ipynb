```python
#@title POS-taging
 #----POS-taging method
def Pos_taging(line):
  tokens = nltk.word_tokenize(line)
 #print(tokens)
  tagged = nltk.pos_tag(tokens)
 #print(tagged)
  return tagged
```

```python
#@title Save { form-width: "10%" }
#-----save result in xlsx format
def save(name):
  workbook = xlsxwriter.Workbook('../data/DataLast/' + name) 
  worksheet = workbook.add_worksheet("HandResult") 

  col = 0
  worksheet.write(0, col, 'requairment') 
  worksheet.write(0, col + 1, 'paper')
  worksheet.write(0, col + 2, 'Subjective')
  worksheet.write(0, col + 3, 'AmbigAdjAdv')
  worksheet.write(0, col + 4, 'Loophole')
  worksheet.write(0, col + 5, 'Open-ended,nonVarify')
  worksheet.write(0, col + 6, 'Superlative')
  worksheet.write(0, col + 7, 'Comparative')
  worksheet.write(0, col + 8, 'Negative')
  worksheet.write(0, col + 9, 'VaguePronoun')
  worksheet.write(0, col + 10, 'Uncertain')
  worksheet.write(0, col + 11, 'Polysemy')
  row = 1
 
  for reqRow, score, score1, score2, score3, score4, score5, score6, score7, score8, score9, score10 in (matrix): 
    worksheet.write(row, col, reqRow) 
    worksheet.write(row, col + 1, score)
    worksheet.write(row, col + 2, score1)
    worksheet.write(row, col + 3, score2)
    worksheet.write(row, col + 4, score3)
    worksheet.write(row, col + 5, score4)
    worksheet.write(row, col + 6, score5)
    worksheet.write(row, col + 7, score6)
    worksheet.write(row, col + 8, score7)
    worksheet.write(row, col + 9, score8)
    worksheet.write(row, col + 10, score9)
    worksheet.write(row, col + 11, score10)
     
    row += 1
  
  workbook.close()
```
