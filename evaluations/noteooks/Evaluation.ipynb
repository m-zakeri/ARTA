```
#@title Precision { form-width: "10%" }
#----calculate precision
def Precision(column,DicType):
  TP=0
  FP=0
  for row in range(687):#LabeledDataset.shape[0]-1):    # range = 0 - rows of LabeledDataset
    LabeledSmell=LabeledDataset.iat[row,column]
    DicSmell=DicType.iat[row,column]
    if (LabeledSmell != '-' and DicSmell != 0):    # != means there is data(smell) in that cell 
      TP=TP+1
      print(row, LabeledSmell, DicSmell)
    elif (LabeledSmell == '-' and DicSmell != 0):   # == meand there is no data(smell) in that cell
      FP=FP+1
      #print(row, LabeledSmell, DicSmell)
  if TP==0: return 1
  precision= TP/(TP+FP)
  print("Number of correct finding smells: " + str(TP), "    Number of finding that are not smells in experts opinions: " +str(FP))

  return "Precision: "+str(precision)
```

```
#@title Recall { form-width: "10%" }
#----calculate recall
def Recall(column,DicType):
  TP=0
  FN=0
  for row in range(687):#LabeledDataset.shape[0]-1):    # range = 0 - rows of LabeledDataset
    LabeledSmell=LabeledDataset.iat[row,column]
    DicSmell=DicType.iat[row,column]
    if (LabeledSmell != '-' and DicSmell != 0):
      TP=TP+1
    elif (LabeledSmell != '-' and DicSmell == 0):
      FN=FN+1
      #print(row, LabeledSmell, DicSmell)
  if TP==0 and FN==0 : return 1
  recall = TP/(TP+FN) 
  print("Number of correct finding smells: " + str(TP), "    Number of corrcet smells that have not been found: " +str(FN))
  return  "Recall: "+str(recall)
```
