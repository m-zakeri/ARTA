```python
#-----calculate similarity for frequent words (get vector) by gensim w2v
frqList = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Frequent.p", "rb" ) )

for i in frqList:
  r=wikiWord2Vec.similarity(i[0],'_'+str(i[0]))
  print(i[0], r)
```

```python
#------ calculate cosine similarity
from scipy import spatial
Similarity=[]
with open("/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Aaaat.txt", 'w' , encoding='utf-8') as f:

  frqList = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Frequent.p", "rb" ) )
  for i in frqList:
  #r=wikiWord2Vec.similarity(i[0],'_'+str(i[0]))
    try:
      a = wikiWord2Vec[i[0]] 
      b = wikiWord2Vec['_'+str(i[0])]
      c = 1 - spatial.distance.cosine(a, b)
    except: 
      c = 1
    Similarity.append([i[0],c])
    f.writelines(str(i[0])+": "+str(c)+"\n")

pickle.dump(Similarity, open("/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Aaaat.p","wb"))

print(Similarity)
```

```python
#-------- calculate avrage similarity
avrage=[]
SS = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Li.p", "rb" ) )
Lw = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Li.p", "rb" ) )
Ec = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Li.p", "rb" ) )
Ci = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Li.p", "rb" ) )
At = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Li.p", "rb" ) )
Li = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Li.p", "rb" ) )
EE = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/EE.p", "rb" ) )
ME = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Me.p", "rb" ) )
Sp = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Sp.p", "rb" ) )
Me = pickle.load( open( "/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/Me.p", "rb" ) )
for i in range(1000):
  av = (SS[i][1] + Lw[i][1] + Ec[i][1] + Ci[i][1] + At[i][1] + Li[i][1] + EE[i][1] + ME[i][1] + Sp[i][1] + Me[i][1]) / 10
  avrage.append([av, Li[i][0]])
avrage.sort()

pickle.dump(avrage, open("/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/avrage.p","wb"))

with open("/content/drive/My Drive/Colab Notebooks/SE-Req/Data/Result/avrage.txt", 'w' , encoding='utf-8') as f:
  f.writelines(str(avrage))
avrage
```
