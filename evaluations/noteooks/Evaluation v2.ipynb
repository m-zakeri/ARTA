```python
def FP(column,DicType):
  FP=0
  for row in range(reqNum):#LabeledDataset.shape[0]-1):    # range = 0 - rows of LabeledDataset
    l=[]
    l.clear()
    LabeledSmell=LabeledDataset.iat[row,column]
    DicSmell=DicType.iat[row,column]
    
    if DicSmell!= '-':                                 # != means there is data(smell) in that cell
        for j in LabeledSmell.split('*'):
            l.append(lem.lemmatize(j))
        for i in DicSmell.split('*'):
            if i not in l:
            #print(l)
                FP=FP+1
               # print(row, LabeledSmell, DicSmell,"FP")
  return FP
```

```python
def FN(column,DicType):
  FN=0
  for row in range(reqNum):#LabeledDataset.shape[0]-1):    # range = 0 - rows of LabeledDataset
    l=[]
    l.clear()
    LabeledSmell=LabeledDataset.iat[row,column]
    DicSmell=DicType.iat[row,column]
    
    if LabeledSmell!= '-':                                 # != means there is data(smell) in that cell
#         print('row:',row, 'DicSmell:', DicSmell, 'Labeled_smell:', LabeledSmell)
        for j in LabeledSmell.split('*'):
            l=lem.lemmatize(j)
            if l not in DicSmell.split('*'):
                print(l)
                FN=FN+1
               # print(row, LabeledSmell, DicSmell,"FN")
  return FN
```

```python
def Precision4(column,DicType):
    tp=TP(column,DicType)
    fp=FP(column,DicType)
    if tp==0: return 1 #"TP=0"

    return tp/(tp+fp)
```

```python
def Recall4(column,DicType):
    tp=TP(column,DicType)
    fn=FN(column,DicType)
    if tp==0 and fn==0 : return 1 #"TP and FN == 0"

    return tp/(tp+fn)
```

```python
#----calculate F-measure
def F_measure2(Presicion,Recall):
    return 2*(Presicion*Recall)/(Presicion+Recall)
```

```python
#-----calculate False Positive Rate    FP/TN+FP
def FPR(column,DicType):
    fp=FP(column,DicType)
    tn=TN(column,DicType)
    if fp==0 and tn==0 : return 1 #"TP and FN == 0"

    return fp/(tn+fp)
```
